calculate answer given task: 

```python
# estimate ratio

def calculate_ratio(vec): 
    v = vec[0] # values 
    i = month_to_num(vec[1]) # index 
    r = v[i]/v[i+6]
    return round(r, 2)
```

```python
def month_to_num(string):
    '''given string of a month, return its index (0-indexing)'''
    if string == "January":
        return 0
    elif string == "February": 
        return 1
    elif string == "March":
        return 2
    elif string == "April": 
        return 3
    elif string == "May":
        return 4
    elif string == "June":
        return 5 
```

```python
def calculate_gap(vec): 
    v = vec[0] # values 
    i = month_to_num(vec[1]) # index 
    g = abs(v[i] - v[i+6])
    return g
```

```python
# compare gap 

def compare_gap(vec): 
    v, indices = vec[0], vec[1]
    i, j = indices[0], indices[1]
    g1 = abs(v[i] - v[i+6])
    g2 = abs(v[j]-v[j+6])
    if g1 > g2:
        return 1
    elif g1 < g2: 
        return 2
```

```python
# compare ratio

def compare_ratio(vec): 
    v, indices = vec[0], vec[1]
    i, j = indices[0], indices[1]
    r1 = round(v[i]/v[i+6], 2)
    r2 = round(v[j]/v[j+6], 2)
    if r1 > r2: 
        return 1
    elif r1 < r2: 
        return 2
    else: 
        return None
```

```python
# estimate gap 
est_gap_ans = df[['values', 'est_month']].apply(calculate_gap, axis=1)
est_ratio_ans = df[['values', 'est_month']].apply(calculate_ratio, axis=1)
comp_gap_ans = df[['values', 'indices']].apply(compare_gap, axis=1)
comp_ratio_ans = df[['values', 'indices']].apply(compare_ratio, axis=1)
```

```python
# save answers into the ans_dataframe

ans_df = pd.DataFrame()
ans_df['est_gap'] = est_gap_ans
ans_df['est_ratio'] = est_ratio_ans
ans_df['comp_gap'] = comp_gap_ans
ans_df['comp_ratio'] = comp_ratio_ans
```

```python
ans_df.to_csv("0_ans.csv")
```
